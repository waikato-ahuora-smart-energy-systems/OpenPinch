## Study question and data

**Study question:** How are validated results turned into reviewable tables, figures, workbooks, and dashboards?

The sample data is packaged with OpenPinch, so the notebook runs without path setup. Read the named inputs and assumptions before substituting plant data.

## Step 1: Prepare reportable cached results

Run this cell once, then inspect its named outputs. Arguments on the method call apply to this analysis; stored configuration is only the fallback when an argument is omitted.

```python
from pathlib import Path

from OpenPinch import PinchProblem, PinchWorkspace

problem = PinchProblem("basic_pinch.json", project_name="Site")
problem.update_options({"THERMAL_DT_CONT": 10.0})
problem.set_dt_cont_multiplier(1.0)
problem.target.all_heat_integration()
problem_data = problem.problem_data
problem_filepath = problem.problem_filepath
cached_results = problem.results
metrics = problem.metrics()
summary = problem.summary_frame()
report = problem.report()
summary
```

## Step 2: Export problem outputs

Run this cell once, then inspect its named outputs. Arguments on the method call apply to this analysis; stored configuration is only the fallback when an argument is omitted.

```python
output_dir = Path("openpinch-results")
input_payload = problem.to_problem_json()
reloaded = PinchProblem(project_name="Site")
reloaded.load("basic_pinch.json")
net_loads = problem.plot.net_load_profiles()
graph_data = problem.plot.data()
catalog = problem.plot.catalog()
plot_paths = problem.plot.export(
    output_dir / "plots", plot=problem.plot.grand_composite_curve
)
gallery = problem.plot.export_gallery(output_dir / "gallery")
workbook = problem.export_excel(output_dir / "tables")
```

## Step 3: Publish active-workspace outputs

Run this cell once, then inspect its named outputs. Arguments on the method call apply to this analysis; stored configuration is only the fallback when an argument is omitted.

```python
workspace = PinchWorkspace(
    "basic_pinch.json", project_name="Site"
)
workspace.target.all_heat_integration()
workspace.components.inventory
workspace.design
workspace.plot.catalog()
workspace.export_excel(output_dir / "workspace-tables")
workspace.show_dashboard()
# Explicit interactive side effect; run after targeting.
dashboard = problem.show_dashboard()
```

## Review the result

Inspect the cached summary and net-load view before reviewing publication outputs; verify method, scope, units, and case identity before sharing files.

```python
from IPython.display import display

publication_outputs = {
    "plot paths": plot_paths,
    "gallery": gallery,
    "problem workbook": workbook,
}
display(summary)
display(net_loads)
display(catalog)
display(publication_outputs)
```
