## Study question and data

**Study question:** How can a prepared case set be saved, restored, and validated without rerunning analysis implicitly?

The sample data is packaged with OpenPinch, so the notebook runs without path setup. Read the named inputs and assumptions before substituting plant data.

## Step 1: Inspect the active case

Run this cell once, then inspect its named outputs. Arguments on the method call apply to this analysis; stored configuration is only the fallback when an argument is omitted.

```python
from pathlib import Path

from OpenPinch import PinchWorkspace

workspace = PinchWorkspace("basic_pinch.json", project_name="Site")
workspace.target.direct_heat_integration()
active_problem = workspace.case()
problem_data = active_problem.problem_data
problem_filepath = active_problem.problem_filepath
master_zone = active_problem.master_zone
cached_results = workspace.results
summary = workspace.summary_frame()
metrics = workspace.metrics()
report = workspace.report()
summary
```

## Step 2: Persist and restore the workspace

Run this cell once, then inspect its named outputs. Arguments on the method call apply to this analysis; stored configuration is only the fallback when an argument is omitted.

```python
bundle_path = Path("openpinch-workspace.json")
saved = workspace.save_bundle(bundle_path)
restored = PinchWorkspace.load_bundle(saved)
loaded = PinchWorkspace(project_name="Site")
case_input = restored.to_problem_json(case_name="baseline")
validated = restored.validate("baseline")
validation = restored.validation_report("baseline")
case_input
```

## Review the result

Compare the original summary with the restored validation and case payload to confirm that the saved workspace contains a reusable study rather than only a file reference.

```python
from IPython.display import display

display(summary)
display(validation)
display(case_input)
```
