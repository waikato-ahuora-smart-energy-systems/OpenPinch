## Study question and data

**Study question:** How does a process engineer compare named design assumptions without mixing their state?

The sample data is packaged with OpenPinch, so the notebook runs without path setup. Read the named inputs and assumptions before substituting plant data.

## Step 1: Create and compare two cases

Run this cell once, then inspect its named outputs. Arguments on the method call apply to this analysis; stored configuration is only the fallback when an argument is omitted.

```python
from OpenPinch import PinchWorkspace

workspace = PinchWorkspace("crude_preheat_train.json", project_name="Crude Site")
baseline = workspace.case("baseline")
baseline.target.direct_heat_integration()
tight_dt = workspace.scenario("tight_dt", dt_cont_multiplier=0.75)
tight_dt.target.direct_heat_integration()
case_comparison = workspace.compare_cases("baseline", "tight_dt")
baseline_comparison = baseline.compare_to(tight_dt)
```

## Step 2: Batch, inspect, and select cases

Run this cell once, then inspect its named outputs. Arguments on the method call apply to this analysis; stored configuration is only the fallback when an argument is omitted.

```python
batch = workspace.cases(["baseline", "tight_dt"])
batch_results = batch.target.direct_heat_integration()
batch_summaries = batch.summary_frames()
batch_metrics = batch.metrics()
batch_reports = batch.reports()
batch_errors = batch_results.errors
workspace.use_case("tight_dt")
baseline.update_options({"THERMAL_DT_CONT": 10.0})
baseline.set_dt_cont_multiplier(1.0)
baseline.config
workspace.update_options({"THERMAL_DT_CONT": 12.0})
workspace.set_dt_cont_multiplier(1.0)
workspace.config
workspace.project_name
workspace.baseline_name
workspace.active_case_name
workspace.list_cases()
```

## Review the result

Use the case comparison for the engineering change and the batch summaries for a consistent review across every selected scenario.

```python
from IPython.display import display

display(case_comparison)
display(baseline_comparison)
display(batch_summaries)
```
