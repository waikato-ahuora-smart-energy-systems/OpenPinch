## Study question and data

**Study question:** How do energy targets translate into exchanger area, cost, and thermodynamic quality?

The sample data is packaged with OpenPinch, so the notebook runs without path setup. Read the named inputs and assumptions before substituting plant data.

## Step 1: Estimate area and cost

Run this cell once, then inspect its named outputs. Arguments on the method call apply to this analysis; stored configuration is only the fallback when an argument is omitted.

```python
from OpenPinch import PinchProblem

problem = PinchProblem("basic_pinch.json", project_name="Site")
area_cost = problem.target.heat_exchanger_area_and_cost()
area_cost_summary = problem.summary_frame()
area_cost_summary
```

## Step 2: Add exergy and period views

Run this cell once, then inspect its named outputs. Arguments on the method call apply to this analysis; stored configuration is only the fallback when an argument is omitted.

```python
base_target = problem.target.direct_heat_integration()
exergy = problem.target.exergy(base_target=base_target)
period_area_cost = problem.target.all_periods.heat_exchanger_area_and_cost()
period_exergy = problem.target.all_periods.exergy()
exergy_curve = problem.plot.exergetic_grand_composite_curve()
exergy_loads = problem.plot.exergetic_net_load_profiles()
```

## Review the result

Read the area and cost summary together with the exergy views to avoid selecting a design from energy targets alone.

```python
from IPython.display import display

display(area_cost_summary)
display(exergy_curve)
display(exergy_loads)
```

## Interpret the result

Treat area and exergy as complementary screens: area exposes transfer difficulty, while exergy exposes quality loss.
