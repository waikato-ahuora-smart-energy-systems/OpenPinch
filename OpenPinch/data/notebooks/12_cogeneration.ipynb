## Study question and data

**Study question:** How sensitive is cogeneration potential to the turbine model used for screening?

The sample data is packaged with OpenPinch, so the notebook runs without path setup. Read the named inputs and assumptions before substituting plant data.

## Step 1: Establish the cogeneration reference

Run this cell once, then inspect its named outputs. Arguments on the method call apply to this analysis; stored configuration is only the fallback when an argument is omitted.

```python
from OpenPinch import PinchProblem

problem = PinchProblem("pulp_mill.json", project_name="Site")
problem.target.all_heat_integration()
base = problem.target.total_site_heat_integration()
default = problem.target.cogeneration(base_target=base)
cogeneration_summary = problem.summary_frame()
```

## Step 2: Compare named turbine models

Run this cell once, then inspect its named outputs. Arguments on the method call apply to this analysis; stored configuration is only the fallback when an argument is omitted.

```python
sun_smith = problem.target.sun_smith_cogeneration(base_target=base)
varbanov = problem.target.varbanov_cogeneration(base_target=base)
isentropic = problem.target.isentropic_cogeneration(
    efficiency=0.8, base_target=base
)
```

## Review the result

Compare the common process summary with the alternative cogeneration results to see how utility conditions and turbine assumptions change recoverable power.

```python
from IPython.display import display

cogeneration_methods = {
    "default": default,
    "Sun-Smith": sun_smith,
    "Varbanov": varbanov,
    "isentropic": isentropic,
}
display(cogeneration_summary)
display(cogeneration_methods)
```

## Interpret the result

Compare models against the same Total Site base target and state efficiency assumptions beside the reported power potential.
