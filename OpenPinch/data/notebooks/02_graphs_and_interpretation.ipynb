# Graphs and Interpretation

Inspect the available graph outputs, build the main figures, and connect each
graph type to the decision it supports.


```python
from pathlib import Path
from tempfile import TemporaryDirectory

from OpenPinch import PinchProblem
from OpenPinch.resources import copy_sample_case

workspace = TemporaryDirectory()
work_dir = Path(workspace.name)
case_path = copy_sample_case("basic_pinch.json", work_dir / "basic_pinch.json")
problem = PinchProblem(problem_filepath=case_path)
problem.run()
problem.graph_catalog()
```

## What the main graphs are for

- Composite curves show the overall overlap between process sources and sinks.
- Shifted composite curves show that overlap after applying the minimum
  approach temperature.
- Grand composite curves are usually the first graph to inspect for utility
  selection and heat-pump opportunity identification.
- The `Grand Composite Curve with Heat Pump` view is useful when reasoning
  about where a temperature lift might be inserted.


```python
composite_curve = problem.plot_composite_curve()
grand_composite_curve = problem.plot_grand_composite_curve()
heat_pump_gcc = problem.plot(graph_type="Grand Composite Curve with Heat Pump")
[
    composite_curve.layout.title.text,
    grand_composite_curve.layout.title.text,
    heat_pump_gcc.layout.title.text,
]
```

## What to inspect first

Use the grand composite curve when you need to answer utility-level questions.
Use the composite curves when you need to explain the overall heat-source and
heat-sink overlap. Always check the utility targets as well as the graph shape.
