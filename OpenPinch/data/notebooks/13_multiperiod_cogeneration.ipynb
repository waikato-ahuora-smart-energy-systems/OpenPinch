## Study question and data

**Study question:** How does cogeneration potential vary with the site's operating periods?

The sample data is packaged with OpenPinch, so the notebook runs without path setup. Read the named inputs and assumptions before substituting plant data.

## Step 1: Run the period reference and default model

Run this cell once, then inspect its named outputs. Arguments on the method call apply to this analysis; stored configuration is only the fallback when an argument is omitted.

```python
from OpenPinch import PinchProblem

problem = PinchProblem("zonal_site_multiperiod.json", project_name="Site")
problem.target.all_periods.all_heat_integration()
period_cogeneration = problem.target.all_periods.cogeneration()
weighted = problem.summary_frame(include_weighted_average=True)
weighted
```

## Step 2: Compare period turbine models

Run this cell once, then inspect its named outputs. Arguments on the method call apply to this analysis; stored configuration is only the fallback when an argument is omitted.

```python
sun_smith = problem.target.all_periods.sun_smith_cogeneration()
varbanov = problem.target.all_periods.varbanov_cogeneration()
isentropic = problem.target.all_periods.isentropic_cogeneration(efficiency=0.8)
list(problem.period_results)
```

## Review the result

Review the weighted table with all period method results to identify which operating conditions drive annual cogeneration value.

```python
from IPython.display import display

period_cogeneration_methods = {
    "default": period_cogeneration,
    "Sun-Smith": sun_smith,
    "Varbanov": varbanov,
    "isentropic": isentropic,
}
display(weighted)
display(period_cogeneration_methods)
```

## Interpret the result

Check the period with the lowest viable steam flow as well as the weighted annual opportunity.
