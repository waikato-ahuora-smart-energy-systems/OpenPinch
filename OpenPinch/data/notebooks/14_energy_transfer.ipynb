**Study question:** Where does heat cross zone boundaries, and which transfers merit utility-system or project attention?

## Step 1: Quantify site energy transfer

Run this cell once, then inspect its named outputs. Arguments on the method call apply to this analysis; stored configuration is only the fallback when an argument is omitted.

```python
from OpenPinch import PinchProblem

site = PinchProblem("pulp_mill.json", project_name="Site")
site.target.all_heat_integration()
base = site.target.total_site_heat_integration()
energy_transfer = site.target.energy_transfer(base_target=base)
transfer_summary = site.summary_frame()
```

## Step 2: Visualize periods and compare a focused zone

Run this cell once, then inspect its named outputs. Arguments on the method call apply to this analysis; stored configuration is only the fallback when an argument is omitted.

```python
diagram = site.plot.energy_transfer_diagram()
period_transfers = site.target.all_periods.energy_transfer()
bleaching = PinchProblem("pulp_mill.json", project_name="Site")
bleaching.target.direct_heat_integration(zone="Bleaching")
comparison = site.compare_to(
    bleaching,
    scope="Site/Bleaching",
    zone_type="Process Zone",
    integration_type="Process",
    target_method="Heat Exchange",
)
```

## Review the result

Use the transfer summary and diagram to screen magnitude and temperature, then use the focused comparison to challenge whether the connection is physically credible.

```python
from IPython.display import display

display(transfer_summary)
display(diagram)
display(comparison)
```

## Interpret the result

Interpret transfers with their source and sink zones; a large transfer is an opportunity only if the physical connection is credible.
