## Study question and data

**Study question:** Which ranked heat-exchanger network best balances recovery, utilities, area, and cost?

The sample data is packaged with OpenPinch, so the notebook runs without path setup. Read the named inputs and assumptions before substituting plant data.

## Step 1: Generate ranked network candidates

Run this cell once, then inspect its named outputs. Arguments on the method call apply to this analysis; stored configuration is only the fallback when an argument is omitted.

```python
from OpenPinch import PinchProblem

problem = PinchProblem(
    "Four-stream-Yee-and-Grossmann-1990-1.json", project_name="Four Stream"
)
design = problem.design.heat_exchanger_network(
    approach_temperatures=[14.0], stages=[1], best_solutions=1
)
ranked_networks = design.top(1)
```

## Step 2: Inspect and serialize a selected network

Run this cell once, then inspect its named outputs. Arguments on the method call apply to this analysis; stored configuration is only the fallback when an argument is omitted.

```python
selected = design.network(rank=1)
selected_property = design.selected_network
grid = design.grid(rank=1)
high_pressure_steam = design.utility("HP Steam")
metrics = {
    "recovery": design.total_heat_recovery,
    "hot utility": design.total_hot_utility,
    "cold utility": design.total_cold_utility,
}
serialized = design.result.model_dump(mode="json")
metrics
```

## Review the result

Review ranked candidates, recovery and utility metrics, and the selected grid together; objective rank alone is not a complete network-selection basis.

```python
from IPython.display import display

display(ranked_networks)
display(metrics)
display(grid)
```

Review several feasible networks and the exchanger grid; never select rank one from objective value alone.
