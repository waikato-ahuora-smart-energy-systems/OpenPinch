## Study question and data

**Study question:** What are the minimum heating and cooling duties, and which curves explain them?

The sample data is packaged with OpenPinch, so the notebook runs without path setup. Read the named inputs and assumptions before substituting plant data.

## Step 1: Prepare and solve one case

Run this cell once, then inspect its named outputs. Arguments on the method call apply to this analysis; stored configuration is only the fallback when an argument is omitted.

```python
from OpenPinch import PinchProblem

problem = PinchProblem("basic_pinch.json", project_name="Site")
validated = problem.validate()
validation = problem.validation_report()
target = problem.target.all_heat_integration()
cached_results = problem.results
summary = problem.summary_frame()
summary
```

## Step 2: Inspect cached engineering outputs

Run this cell once, then inspect its named outputs. Arguments on the method call apply to this analysis; stored configuration is only the fallback when an argument is omitted.

```python
metrics = problem.metrics()
report = problem.report()
available_plots = problem.plot.catalog()
graph_data = problem.plot.data()
available_plots
```

## Step 3: Build the core diagnostic curves

Run this cell once, then inspect its named outputs. Arguments on the method call apply to this analysis; stored configuration is only the fallback when an argument is omitted.

```python
composite = problem.plot.composite_curve()
shifted = problem.plot.shifted_composite_curve()
balanced = problem.plot.balanced_composite_curve()
grand = problem.plot.grand_composite_curve()
real_grand = problem.plot.real_grand_composite_curve()
```

## Review the result

Review the target table first, then use the composite and grand composite curves to connect the utility targets to the remaining heat surplus and deficit.

```python
from IPython.display import display

display(summary)
display(composite)
display(grand)
```

Check the summary duties first, then use the composite and grand-composite curves to locate the pinch and utility opportunities.
